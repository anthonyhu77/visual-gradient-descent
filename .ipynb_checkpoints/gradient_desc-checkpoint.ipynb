```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
```

```python
#100 data points

def generateRanLine(data_points):
    m = random.random()
    b = np.random.randint(data_points/4, size=data_points)
    x = np.random.randint(100, size=data_points)
    y = m*x + b
    
    
    plt.scatter(x,y) #Plotting the initialized data
    plt.show()
    
    return y, x, m, b
    
    
Y, X, true_m, true_b  = generateRanLine(100)
    
```

```python
def gradient_descent(X,Y,m=0,b=0,L=0.0001,epochs=1000): #Performing Gradient Descent 
    pred_y_epochs = [None] * epochs #Alocate space for predict y values every epoch
    n = float(len(X))
    
    for i in range(epochs): 
        Y_pred = m*X + b  # The current predicted value of Y
        pred_y_epochs[i] = Y_pred
        D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
        D_b = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
        m = m - L * D_m  # Update m
        b = b - L * D_b  # Update c
        
    print (m, b)
    return m, b, pred_y_epochs

pred_m, pred_b, pred_y_epochs = gradient_descent(X,Y) 
```

```python
fig, ax = plt.subplots()
ax.scatter(X, Y) 
ax.plot([min(X), max(X)], [min(pred_y_epochs[-1]), max(pred_y_epochs[-1])], color='red')  # regression line
```

```python
fig, ax = plt.subplots()
ax.scatter(X, Y) 
line, = ax.plot([min(X), max(X)], [min(pred_y_epochs[0]), max(pred_y_epochs[0])], color='red')  # regression line
```

```python
def update(i):
    label = "epoch {0}".format(i+1)
    line.set_ydata(X+i)
    ax.set_xlabel(label)
    return line, ax
```

```python
fig, ax = plt.subplots()
x = np.arange(0,20,0.1)
ax.scatter(x, x + np.random.normal(0,3.0,len(x)))
line, = ax.plot(x, x-5,'r-', linewidth=2)

def update(i):
    label = "epoch {0}".format(i+1)
    line.set_ydata(x - 5 + i)
    ax.set_xlabel(label)
    return line, ax


anim = FuncAnimation(fig, update, repeat = True, frames = np.arange(0,100), interval = 500)
plt.show()
```
