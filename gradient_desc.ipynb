```python
#Import 
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
#100 data points
def generateRanLine(data_points):
    m = random.random()
    b = np.random.normal(0, 5, size=data_points)
    x = np.random.randint(100, size=data_points)
    y = m*x + b 
    plt.scatter(x,y) #Plotting the initialized data
    plt.show()
    
    return y, x, m, b
    
    
Y, X, true_m, true_b  = generateRanLine(100)
    
```

```python
def gradient_descent(X,Y,m=np.random.normal(0,1),b=np.random.normal(0,1),L=0.000001,epochs=10000): #Performing Gradient Descent 
    pred_y_epochs = [None] * epochs #Alocate space for predict y values every epoch
    n = float(len(X))
    
    for i in range(epochs): 
        Y_pred = m*X + b  # The current predicted value of Y
        pred_y_epochs[i] = Y_pred
        D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
        D_b = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
        m = m - L * D_m  # Update m
        b = b - L * D_b  # Update c
        
    print (m, b)
    return m, b, pred_y_epochs

pred_m, pred_b, pred_y_epochs = gradient_descent(X,Y) 
```
